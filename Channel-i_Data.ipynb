```python
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
```

```python
path=r"C:\Selenium\IEDriverServer.exe"
browser = webdriver.Ie(path)
# Replace <webdriver_path> with path to your webdriver
browser.get('https://channeli.in/')
sign_in = WebDriverWait(browser, 15).until(EC.presence_of_element_located((By.ID, 'login')))
sign_in.click()
WebDriverWait(browser, 15).until(EC.presence_of_element_located((By.ID, 'login_dialog-div')))
browser.switch_to.frame(browser.find_element_by_id('login_dialog-iframe'))
username = WebDriverWait(browser, 1).until(EC.presence_of_element_located((By.NAME, 'username')))
username.send_keys('Enrollment Number') # Replace <username> with your username
password = WebDriverWait(browser, 1).until(EC.presence_of_element_located((By.NAME, 'password')))
password.send_keys('Password') # Replace <password> with your password
submit = browser.find_element_by_id('sign_in_button').click()
```

```python
browser.get('https://channeli.in/placement/results/branch/')
```

```python
branch=[]
selected=[]

rows=browser.find_element_by_tag_name("tbody").find_elements_by_tag_name("tr")

for row in rows:
    x=row.find_elements_by_tag_name("td")
    branch.append(x[2].text + " (" + x[1].text + ")")
    selected.append(x[3].text)

Data=pd.DataFrame({"selected_2016":selected},index=branch)
```

```python
for i in range(2015,2008,-1):
    link="https://channeli.in/placement/results/"+str(i)+"/branch/"
    browser.get(link)
    
    branch=[]
    selected=[]

    rows=browser.find_element_by_tag_name("tbody").find_elements_by_tag_name("tr")

    for row in rows:
        x=row.find_elements_by_tag_name("td")
        branch.append(x[2].text + " (" + x[1].text + ")")
        selected.append(x[3].text)
    
    column_name="selected_"+str(i)
    Data1=pd.DataFrame({column_name:selected},index=branch)

    Data  = pd.concat([Data,Data1],axis=1)
```

```python
Data.to_csv("Branch_wise.csv")
```

```python
#Department_wise
```

```python
browser.get("https://channeli.in/placement/results/2016/department/")

branch=[]
overall_url=[]
total_students=[]
registered_students=[]
placed_students=[]

rows=browser.find_element_by_tag_name("tbody").find_elements_by_tag_name("tr")

for row in rows:
    x=row.find_elements_by_tag_name("td")
    branch.append(x[1].text)
    overall_url.append(x[2].find_element_by_tag_name('a').get_attribute('href'))


for url in overall_url:
    browser.get(url)
    row=browser.find_element_by_tag_name("tbody").find_element_by_tag_name("tr").find_elements_by_tag_name("td")
    total_students.append(row[0].text)
    registered_students.append(row[1].text)
    placed_students.append(row[2].text)

Data=pd.DataFrame({"total_students_2016":total_students,"registered_students_2016":registered_students,"placed_students_2016":placed_students},index=branch)
Data=Data[["total_students_2016","registered_students_2016","placed_students_2016"]]
```

```python
for i in range(2015,2008,-1):
    link="https://channeli.in/placement/results/"+str(i)+"/department/"
    browser.get(link)

    branch=[]
    overall_url=[]
    total_students=[]
    registered_students=[]
    placed_students=[]

    rows=browser.find_element_by_tag_name("tbody").find_elements_by_tag_name("tr")

    for row in rows:
        x=row.find_elements_by_tag_name("td")
        branch.append(x[1].text)
        overall_url.append(x[2].find_element_by_tag_name('a').get_attribute('href'))


    for url in overall_url:
        browser.get(url)
        row=browser.find_element_by_tag_name("tbody").find_element_by_tag_name("tr").find_elements_by_tag_name("td")
        total_students.append(row[0].text)
        registered_students.append(row[1].text)
        placed_students.append(row[2].text)
    
    column1_name="total_students_"+str(i)
    column2_name="registered_students_"+str(i)
    column3_name="placed_students_"+str(i)

    Data1 = pd.DataFrame({column1_name:total_students,column2_name:registered_students,column3_name:placed_students},index=branch)
    Data1 = Data1[[column1_name,column2_name,column3_name]]
    Data  = Data.join(Data1,how="outer")
```
